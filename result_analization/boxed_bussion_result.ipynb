```python
import pandas as pd
import itertools

def get_keys(*args):
    return [i for i in itertools.chain.from_iterable([list(map(lambda x: x[0], data)) for data in args])]

def get_values(*args):
    return [i for i in itertools.chain.from_iterable([list(map(lambda x: x[1], data)) for data in args])]

def create_frame(measures, datas, index=""):
    assert len(measures) == len(datas)
    keys = get_keys(datas[0])
    values = [get_values(d) for d in datas]
    df =  pd.DataFrame(list(zip(*values)), index=keys, columns=measures)
    df.index.rename(index, inplace=True)
    return df.style.highlight_max(color='yellowgreen', subset=measures)
    # return df

e = create_frame(["f1_micro", "f1_macro"],
                 [ltr.get_last_value(boxed_data_bussio, "f1_micro") + ltr.get_last_value(default_data_bussio, "f1_micro")[1:],
                  ltr.get_last_value(boxed_data_bussio, "f1_macro") + ltr.get_last_value(default_data_bussio, "f1_macro")[1:]],
                 "global bussio")
e
```
